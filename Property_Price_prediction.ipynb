```python
import pandas as pd
```

```python
data = pd.read_csv("HousePrices.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.isnull()
```

```python
miss_val = data.columns[data.isnull().any()]
data[miss_val].isnull().sum()
```

```python
data.info()
```

```python
data['LotFrontage'].fillna((data['LotFrontage'].mean()),inplace=True)
```

```python
data['BsmtUnfSF'].fillna(data['BsmtUnfSF'].mean(),inplace=True)
data['LotArea'].fillna(data['LotArea'].mean(),inplace=True)
data['OverallQual'].fillna(data['OverallQual'].mean(),inplace=True)
data['OverallCond'].fillna(data['OverallCond'].mean(),inplace=True)
data['MasVnrArea'].fillna(data['MasVnrArea'].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
x1 = data[['Dwell_Type','LotFrontage', 'LotArea','OverallQual','OverallCond','MasVnrArea',
'BsmtFinSF1','BsmtUnfSF','TotalBsmtSF','1stFlrSF','2ndFlrSF','GrLivArea','GarageArea',
'WoodDeckSF', 'OpenPorchSF']]  #independent
y1 = data['Property_Sale_Price']    #dependent
print(x1,y1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x1,y1)
```

```python
model.score(x1,y1)
```

```python
model.coef_
```

```python
house=model.predict([[70,85,12000,6,8,100,520,430,1024,950,850,1800,700,150,42]])
house
```

```python
from sklearn.model_selection import train_test_split
X1_train, X1_test, y1_train, y1_test = train_test_split(x1, y1, test_size = 0.25, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg_mod = LinearRegression()
lin_reg_mod.fit(X1_train, y1_train)
```

```python
pred = lin_reg_mod.predict(X1_test)
pred
```

```python
import numpy as np
from sklearn import metrics
test_set_r2 = metrics.r2_score(y1_test, pred)
print(test_set_r2)
```
